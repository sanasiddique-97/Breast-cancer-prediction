```python
import os
import pydicom
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Directory containing the main folder with subfolders
dicom_dir = r'E:\cancercases'

# Directory where the JPEG images will be saved
output_dir = r'C:\Users\kenza.chenni\Desktop\acıbademsana\cancercases'
os.makedirs(output_dir, exist_ok=True)  # Create the output directory if it doesn't exist

# Walk through all subfolders in the dicom_dir
for root, _, files in os.walk(dicom_dir):
    for filename in files:
        file_path = os.path.join(root, filename)

        try:
            # Try reading the file as a DICOM file
            dicom_data = pydicom.dcmread(file_path)

            # Extract the pixel array from the DICOM file
            pixel_array = dicom_data.pixel_array

            # Check if the image is grayscale
            if len(pixel_array.shape) == 2:  # Grayscale image
                # Normalize the pixel values to the range 0-255
                pixel_array = (pixel_array - pixel_array.min()) / (pixel_array.max() - pixel_array.min()) * 255
                pixel_array = pixel_array.astype(np.uint8)
            else:
                # Convert to RGB if it's already in a multi-channel format
                pixel_array = pixel_array.astype(np.uint8)

            # Convert the numpy array to a PIL image
            image = Image.fromarray(pixel_array)

            # Define the output path for the JPEG file
            relative_path = os.path.relpath(root, dicom_dir)  # Get relative subfolder path
            output_subfolder = os.path.join(output_dir, relative_path)  # Create the same subfolder structure in the output directory
            os.makedirs(output_subfolder, exist_ok=True)  # Create the subfolder if it doesn't exist

            jpeg_filename = f'{os.path.splitext(filename)[0]}.jpg'
            jpeg_path = os.path.join(output_subfolder, jpeg_filename)

            # Convert the image to JPEG and save it with higher quality
            image.convert('RGB').save(jpeg_path, 'JPEG', quality=95)

            # Optionally display the image (for example, just the first image)
            if filename == os.listdir(root)[0]:
                plt.imshow(pixel_array, cmap='gray')
                plt.show()

            print(f'Converted {filename} to {jpeg_filename} in {output_subfolder}')

        except Exception as e:
            # If the file is not a valid DICOM file, it will raise an exception
            print(f'Failed to read {filename} as DICOM: {e}')
```
