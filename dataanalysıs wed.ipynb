```python
import pandas as pd

# Assuming you already have the DataFrame 'datacombined_df_0'

# List of columns to check for uniqueness
columns_to_check = [
   'InputFileName',
       'StudyDate', 'StudyTime', 'PatientID.x'
]

# Loop through combinations of columns and check for uniqueness
unique_columns = []

for col in columns_to_check:
    # Check uniqueness based on a single column
    unique_rows_count = datacombined_df_1[col].nunique()
    total_rows_count = len(datacombined_df_1)

    if unique_rows_count == total_rows_count:
        unique_columns.append(col)

# Display the columns that can be used for uniqueness
print(f"Columns that can be used to find uniqueness: {unique_columns}")
```
